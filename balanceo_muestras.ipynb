```python
from google.colab import files
import pandas as pd

# Subir archivo
uploaded = files.upload()

# Leer el archivo Excel
for file_name in uploaded.keys():
    df = pd.read_excel(file_name)
    print(f'{file_name} cargado y leído en un DataFrame')

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
import pandas as pd

col = 'SEGMENTO_ACTUAL'

# Filtrar el DataFrame para el segmento 'ALDIA'
df1 = df[df['Categoria'] == 'Champion 70%'][[col]]
df2 = df[df['Categoria'] != 'Champion 70%'][[col]]

# Identificar las columnas de tipo texto
text_columns = df1.select_dtypes(include='object').columns

# Lista para almacenar los DataFrames resultantes
result_list = []

# Función para calcular proporciones
def calculate_proportions(df, col):
    proportions = df[col].value_counts(normalize=True).reset_index()
    proportions.columns = [col, f'{col}_proportion']
    return proportions

# Procesar cada columna de tipo texto
for col in text_columns:
    # Calcular proporciones para df1 y df2
    proportions_df1 = calculate_proportions(df1, col)
    proportions_df2 = calculate_proportions(df2, col)

    # Unir las proporciones de ambos DataFrames
    merged_proportions = proportions_df1.merge(proportions_df2, on=col, how='outer', suffixes=('_Champion', '_Challenger')).fillna(0)

    # Agregar la columna 'categoria' con el nombre de la columna
    #merged_proportions['categoria'] = col

    # Agregar el DataFrame resultante a la lista
    result_list.append(merged_proportions)

# Concatenar todos los resultados en un único DataFrame final
final_df = pd.concat(result_list, ignore_index=True)

# Mostrar el DataFrame final
display(final_df)
```
